Chat Templte  
参考连接: https://huggingface.co/docs/transformers/chat_templating#chat-templates

所有的因果语言模型(causal LMs(由先前输入(上文信息)预测下一个Token)), 本质上都是将文本序列作为输入来处理的。在庞大的文本数据集上进行“预训练”(Pre-Training)，从而生成一个“基础模型”(base model)。 之后进行微调(fine-tuned),以消息序列形式呈现的数据来对模型进行进一步训练。

Chat Template 是“排版”，Tokenizer 是“翻译”。排版永远在翻译之前。

1. 创建一个List of Dicts
message = [
    {"role": "system", "context":"你是AI助手。...."}
    {"role":"uesr", "context":"你好！......."}
] 

2. Chat Template 拼接后的"带控制字符的文本"
<|im_start|>system
你是AI助手。<|im_end|>
<|im_start|>user
你好！<|im_end|>
<|im_start|>assistant

3. 将2 中的长文本, 划分为一个个词根, 并转换为vocab 中的索引(Token IDs)
  接收到的文本 ---> Tensor [312313, 1321313, 1312313,] (数字表示某个词的id, (数字随便写的,不代表真正的Token id))


tokenizer.apply_chat_template(messages) 直接完成2 3步骤. 

```python
from transformers import AutoTokenizer

# HF 官方使用 Mistral 和 Zephyr 模型进行演示, 一开始并未下载对应model 使用已下载的Qwen 进行演示
tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen2.5-0.5B-Instruct", local_files_only=True)
chat = [
  {"role": "user", "content": "Hello, how are you?"},
  {"role": "assistant", "content": "I'm doing great. How can I help you today?"},
  {"role": "user", "content": "I'd like to show off how chat templating works!"},
]

tokenizer.apply_chat_template(chat, tokenize=False)
# tokenizer.apply_chat_template(chat, tokenize=True) 直接将chat 文本转换为 token id

```

在 m01.py 中 直接使用 tokenizer() 或是 tokenizer.ecode() 直接将一个seq 编码为 Token (非对话场景)
常见任务场景 文本分类、情感分析. 亦或者是 纯文本续写任务. 简单理解为 输入一段纯文本, 需要将其编码.


让模型理解对话的上下文和轮次结构, 并且需要处理多轮对话或带有明确角色的交互 情况下 使用 tokenizer.apply_chat_template
